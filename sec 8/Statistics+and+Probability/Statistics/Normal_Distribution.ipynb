```python
from scipy.stats import norm
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
normal = norm.rvs(size=1000,loc=0,scale=1)

ax = sns.distplot(normal,
                  bins=100,
                  kde=True,
                  color='Green')
ax.set(xlabel='Normal Distribution', ylabel='Frequency')
```
